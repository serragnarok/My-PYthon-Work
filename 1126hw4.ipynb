```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
z_list = []
for y in range(3):
    for x in range(3):
        z = np.cos(x) + np.sin(y)
        z_list.append(z)

z = z_list   
x = np.linspace(0,2,3)
y = np.linspace(0,2,3)      
X,Y = np.meshgrid(x,y)
z = np.mat(z)              
z = np.array(z)
z.shape = (3,3)

plt.figure(figsize=(10,6))

plt.contourf(x,y,z)
plt.contour(x,y,z)
```
